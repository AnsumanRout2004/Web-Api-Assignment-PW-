    Q1: What is an API? Give an example, where an API is used in real life.

    Advantages:
    1. Automation: APIs enable automation by allowing software systems to interact without human intervention.
    2. Efficiency: APIs allow developers to use existing functionalities instead of developing from scratch.
    3. Scalability: APIs facilitate easy expansion and integration of new features.
    4. Security: APIs allow secure data exchange by defining access rules.
    
    Disadvantages:
    1. Security Risks: APIs expose endpoints that can be exploited if not secured properly.

    """

def q3():

    A Web API is a specific type of API that is accessed over the internet using HTTP protocols. It enables communication between web-based applications and services.
    
    Differences:
    1. API: A general term referring to any interface that allows communication between software components.
    2. Web API: A subset of APIs that operate over the web and use HTTP methods like GET, POST, PUT, DELETE.
    """

def q4():
    return """
    Q4: Explain REST and SOAP Architecture. Mention shortcomings of SOAP.
    
    REST (Representational State Transfer):
    - Uses standard HTTP methods (GET, POST, PUT, DELETE) for communication.
    - Data is exchanged in formats like JSON or XML.
    - Stateless architecture improves scalability and flexibility.
    
    SOAP (Simple Object Access Protocol):
    - Uses XML format for message exchange.
    - Relies on strict standards for security and transactions.
    - Operates using the SOAP protocol over HTTP, SMTP, or TCP.
    
    Shortcomings of SOAP:
    - More complex due to extensive standards.
    - Slower due to XML parsing.
    - Less flexible compared to REST.
    """
